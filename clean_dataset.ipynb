```python
employees = pd.read_csv('C:/Users/Hp Pc/queries/company_employee/employees.csv')
dept_emp = pd.read_csv('C:/Users/Hp Pc/queries/company_employee/dept_emp.csv')
department = pd.read_csv('C:/Users/Hp Pc/queries/company_employee/departments.csv')
managers = pd.read_csv('C:/Users/Hp Pc/queries/company_employee/dept_manager.csv')
title = pd.read_csv('C:/Users/Hp Pc/queries/company_employee/titles.csv')
```

MERGE Department Data

```python
department = department.set_index('dept_no', drop=True)
```

```python
department
```

```python
dept_list = department.to_dict(orient='dict')
```

```python
dept_list
```

```python
dept_list['dept_name']
```

```python
dept_dict = dept_list['dept_name']
```

```python
dept_emp['dept_no'].map(dept_dict)
```

```python
dept_emp['dept_no'] = dept_emp['dept_no'].map(dept_dict)
managers['dept_no'] = managers['dept_no'].map(dept_dict)
```

```python
managers.head()
```

```python
managers  = managers.merge(employees[['emp_no', 'first_name', 'last_name']], left_on=['emp_no'], right_on=['emp_no'])
```

```python
managers.to_csv('dept_managers.csv', index=False)
```

MERGE EMPLOYEES DATA

```python
title
```

```python
#convert title data to dictionary

title.set_index('title_id', drop=True, inplace=True)
title_list = title.to_dict(orient='dict')
title_dict = title_list['title']
```

```python
employees['emp_title_id'] = employees['emp_title_id'].map(title_dict)
employees['birth_date'] = pd.to_datetime(employees['birth_date'])
employees['hire_date'] = pd.to_datetime(employees['hire_date'])
```

```python
employees.dtypes
```

```python
#full employee table data

employee_data = employees.merge(salaries, left_on=['emp_no'], right_on=['emp_no'])
employee_data = employee_data.merge(dept_emp, left_on=['emp_no'], right_on=['emp_no'])
```

```python
employee_data.rename(columns = {
                        'emp_title_id': 'emp_level',
                        'dept_no': 'department'
}, inplace=True)
```

```python
employee_data.head(10)
```

```python
employee_data.to_csv('full_employee.csv', index=False)
```
